# Loading white wine dataset, getting the insight, and some visualization


```python
# Loading white wine dataset ######################################################################################

df = pd.read_csv('winequality-white.csv', delimiter = ';')
df.head()
print(df.info())

# visualization ###################################################################################################

# Pearson Corr Heat Map with p_values (%95) 

def corr_sig(df=None):
    p_matrix = np.zeros(shape=(df.shape[1],df.shape[1]))
    for col in df.columns:
        for col2 in df.drop(col,axis=1).columns:
            _ , p = stats.pearsonr(df[col],df[col2])
            p_matrix[df.columns.to_list().index(col),df.columns.to_list().index(col2)] = p
    return p_matrix

p_values = corr_sig(df)
mask = np.invert(np.tril(p_values<0.05))
mask[np.triu_indices_from(mask)] = False

corr = df.corr()
fig, ax = plt.subplots(figsize=(10, 10))

sns.set(font_scale = 1.4)
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, cmap=cmap, mask=mask, annot=True, fmt=".2f", annot_kws={"size":14})
plt.xticks(range(len(corr.columns)), corr.columns, size = 14);
plt.yticks(range(len(corr.columns)), corr.columns, size = 14)
plt.show()

print('\n'"*** p_values (%95): ***"'\n''\n',p_values.round(1))
print('\n'"*** Correlation: ***"'\n''\n',df.corr().round(2))


# Pair Plot for Numeric Data ######################################################################################

sns.pairplot(df,hue='quality') 
```

## Data preparation for classification of two groups of wine qualities, good & bad


```python
df_filter= df[df.quality !=6]
df_filter['target'] = np.where(df_filter['quality'] > 6, 'Good Quality', 'Low Quality')
df_filter['target'] = df_filter['target'].replace(['Good Quality','Low Quality'],[1,0])
df_filter

#sns.set(font_scale = 1.2)
#sns.histplot(df_filter, x="quality", hue="target",kde=True)
#plt.show()

target_count = df_filter.target.value_counts()
print('Low quality:', target_count[0])
print('High quality:', target_count[1])
print('Proportion:', round(target_count[0] / target_count[1], 2), ': 1')

sns.set(font_scale = 1.2)
target_count.plot(kind='bar', title='Count (target)');

df_filter.drop(['target'], axis = 1)

```

## Creating classification pipeline with several models and metrics


```python
X_df = df_filter.drop(['target','quality'], axis = 1).values
y_df= df_filter.target.values.ravel()


classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=3),
    RandomForestClassifier(max_depth=3, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]
names = ["Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

scoring = ['accuracy','roc_auc','f1_micro','recall_micro','precision_micro']
data=[]
for name, clf in zip(names, classifiers):
    pipeline = make_pipeline(StandardScaler(), clf)

    cv_results = cross_validate(pipeline,X_df, y_df,
                                scoring=scoring,cv=10,n_jobs=-1)
    tmp_df=pd.DataFrame(cv_results)
    tmp_df['classifier']=name
    data.append(tmp_df)
data=pd.concat(data)
data=data.groupby('classifier').mean().reset_index()
data.drop(columns=['fit_time','score_time'],inplace=True)
data=data.melt(id_vars=['classifier'],var_name='Scoring')

sns.set(font_scale = 1.4)
g = sns.catplot(data=data,
                x="classifier", y="value",hue="Scoring",kind="bar", height=5, aspect=3)
g.set_xticklabels(rotation=90)
g.set(ylim=(0.0, 1.0))
```

## Balancing data set


```python
# Shuffle the Dataset.
shuffled_df = df_filter.sample(frac=1,random_state=4)

# Put all the quality class in a separate dataset.
high_quality_Wine = shuffled_df.loc[shuffled_df['target'] == 1]

#Randomly select number of low quality observations from the Low-quality (majority class)
low_quality_Wine = shuffled_df.loc[shuffled_df['target'] == 0].sample(n=1060,random_state=4)

# Concatenate both dataframes again
normalized_df = pd.concat([high_quality_Wine, low_quality_Wine])

#plot the dataset after the undersampling
sns.set(font_scale = 1.2)
plt.figure(figsize=(5, 5))
sns.countplot('target', data=normalized_df)
plt.title('Balanced Classes')
plt.show()

```

## Repeating classification pipeline with several models and metrics on the balanced dataset


```python
X_norm = normalized_df.drop(['target', 'quality'], axis = 1).values
y_norm = normalized_df.target.values

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    MLPClassifier(alpha=1, max_iter=1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]
names = ["Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

scoring = ['accuracy','roc_auc','f1_micro','recall_micro','precision_micro']
data=[]
for name, clf in zip(names, classifiers):
    pipeline = make_pipeline(StandardScaler(), clf)

    cv_results = cross_validate(pipeline,X_norm, y_norm,
                                scoring=scoring,cv=5,n_jobs=-1)
    tmp_df=pd.DataFrame(cv_results)
    tmp_df['classifier']=name
    data.append(tmp_df)
data=pd.concat(data)
data=data.groupby('classifier').mean().reset_index()
data.drop(columns=['fit_time','score_time'],inplace=True)
data=data.melt(id_vars=['classifier'],var_name='Scoring')

sns.set(font_scale = 1.4)
g = sns.catplot(data=data,
                x="classifier", y="value",hue="Scoring",kind="bar", height=5, aspect=3)
g.set_xticklabels(rotation=90)
g.set(ylim=(0.0, 1.0))

```
